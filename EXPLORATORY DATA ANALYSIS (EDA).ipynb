# To find the correlation between all these features, I used the 'corr()' function.¶

```python
corr_matrix=open1.corr().round(2)
print(corr_matrix)
print(corr_matrix['UER'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
plt.figure(figsize=(24,24))
sns.heatmap(corr_matrix, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 15},
           xticklabels= open1, yticklabels= open1,
           cmap= 'coolwarm')
```
