# The next step is a dicey one
There'shere's a lot of missing data 'NaN' . This might cause a biased analysis or model. The two possible solutions is to either delete missing data or fill missing data.
We shall go for the latter one, however filling missing data depends on whether it's random or not.Random mising data is best solved by either filling with mean of the total missing columns or filling with next/previous available data(fill pad).
This one is non-Random so we are using imputation by linear regression model.
I couldn't split the data because there's only few data. 
Hence the available data is used to predict the missing data.
The independent variable (x) is Date, while the target variable is the available data.

It's a tedious one though

```python
from sklearn.linear_model import LinearRegression as lr
model=lr()

x1=dft[['Date']][0:2]
y1=dft['PD'][0:2]
pd1=model.fit(x1,y1)
pm1=model.predict(dft[['Date']][2:11])
print(pm1)

x=dft[['Date']][0:4]
y=dft['GO'][0:4]
go=model.fit(x,y)
gm=model.predict(dft[['Date']][4:11])
print(gm)
```

Repeat for other features with missing values.

```python
x2=dft[['Date']][0:5]
y2=dft['HS'][0:5]
hs=model.fit(x2,y2)
hs1=model.predict(dft[['Date']][5:11])
print(hs1)


x3=dft[['Date']][0:6]
y3=dft['HM'][0:6]
hm=model.fit(x3,y3)
hm=model.predict(dft[['Date']][6:11])
print(hm)


x4=dft[['Date']][0:5]
y4=dft['RF'][0:5]
rf=model.fit(x4,y4)
rm=model.predict(dft[['Date']][5:11])
print(rm)

x5=dft[['Date']][0:6]
y5=dft['TP'][0:6]
tp=model.fit(x5,y5)
tm=model.predict(dft[['Date']][6:11])
print(tm)


x6=dft[['Date']][0:9]
y6=dft['MP'][0:9]
mp=model.fit(x6,y6)
mm=model.predict(dft[['Date']][9:11])
print(mm)

x7=dft[['Date']][0:7]
y7=dft['SV'][0:7]
sv=model.fit(x7,y7)
sm=model.predict(dft[['Date']][7:11])
print(sm)


x8=dft[['Date']][0:2]
y8=dft['IE'][0:2]
ie=model.fit(x8,y8)
im=model.predict(dft[['Date']][2:11])
print(im)


x9=dft[['Date']][0:7]
y9=dft['ARC'][0:7]
ars=model.fit(x9,y9)
arm=model.predict(dft[['Date']][7:11])
print(arm)


x10=dft[['Date']][0:2]
y10=dft['PI'][0:2]
pi=model.fit(x10,y10)
pm=model.predict(dft[['Date']][2:11])
print(pm)

```

I was able to predict the missing values in assumption that it follows a linear  pattern. 
However, the result for 'IE' is negative because of few available value, so the slope was very high.

# Lastly, I merged the predicted values with there respective available values using the pandas concat function.
Note the slicing I did and the 'ignore_index'. Without it there will be Error notice.

```python
dt=pd.DataFrame(pm1)
concat=pd.concat([dft['PD'][0:2], dt], ignore_index=True)
dft['PD1']= concat
print(dft['PD'])


dt=pd.DataFrame(gm)
concat=pd.concat([dft['GO'][0:4], dt], ignore_index=True)
dft['GO']= concat
print(dft['GO'])

 
```

Repeat for others

```python
dt=pd.DataFrame(hs1)
concat=pd.concat([dft['HS'][0:5], dt], ignore_index=True)
dft['HS']= concat
 
dt=pd.DataFrame(hm)
concat=pd.concat([dft['HM'][0:6], dt], ignore_index=True)
dft['HM']= concat
 
 
dt=pd.DataFrame(rm)
concat=pd.concat([dft['RF'][0:5], dt], ignore_index=True)
dft['RF']= concat
 
 
dt=pd.DataFrame(tm)
concat=pd.concat([dft['TP'][0:6], dt], ignore_index=True)
dft['TP']= concat
 
dt=pd.DataFrame(mm)
concat=pd.concat([dft['MP'][0:9], dt], ignore_index=True)
dft['MP']= concat
 
 
dt=pd.DataFrame(sm)
concat=pd.concat([dft['SV'][0:7], dt], ignore_index=True)
dft['SV']= concat
 
 
dt=pd.DataFrame(im)
concat=pd.concat([dft['IE'][0:2], dt], ignore_index=True)
dft['IE1']= concat
 
dt=pd.DataFrame(arm)
concat=pd.concat([dft['ARC'][0:7], dt], ignore_index=True)
dft['ARC']= concat
 
 
dt=pd.DataFrame(pm)
concat=pd.concat([dft['PI'][0:2], dt], ignore_index=True)
dft['PI1']= concat
 
 
 
print(dft)
```

Using linear regression for imputation method has its benefits, however, in cases of datasets with few values, it perform woefully because of the high gradient slope. This gives negatives values as seen in (PD, IE & PI).
To solve this problem, impute the missing values with the last minimum vale.

```python
dft['PD']=dft['PD'].fillna(22.7)
dft['IE']=dft['IE'].fillna(0.03)
dft['PI']=dft['PI'].fillna(32.4)
```
